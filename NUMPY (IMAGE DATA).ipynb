```python
fig,ax=plt.subplots(1,3)
ch=['Red','Green','Blue']
for i,axi in enumerate(ax.flat):
    axi.imshow(img[:,:,i],plt.cm.gray)
    axi.set(xticks=[],yticks=[])
    axi.set_xlabel(ch[i],color=ch[i].lower())
```

```python
img2=img.copy()
img2[:,:,2]=200
plt.imshow(img2)
```

```python
img2=img.copy()
img2[400:500,400:500,1]=200
img2[400:1000,400:1000,0]=0
img2[400:1000,400:1000,2]=0
plt.imshow(img2)
```

```python
img2=img.copy()
plt.imshow(img2)
```

```python
idx=np.arange(img2.shape[1])
img3=img2[:,idx[::-1],:]
plt.imshow(img3)
```

```python
plt.imshow(img3[np.arange(img3.shape[0])[::-1],:,:])
```
